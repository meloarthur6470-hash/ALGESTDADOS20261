```python
#1 Uma empresa de médio porte deseja realizar um levantamento sobre a remuneração de seus funcionários. Desenvolva um programa que leia os salários de 10 funcionários, um por vez, e, ao final, informe a média salarial da empresa, o maior salário registrado e quantos funcionários recebem acima de R$ 3.000. Os valores informados são válidos e positivos.
soma = 0
maior = 0
acima_3000 = 0
for i in range(10):
    salario = float(input("digite o salario: "))
    soma += salario
    if salario > maior:
        maior = salario
        if salario > 3000:
            acima_3000 += 1
media = soma / 10
print("A media salarial da empresa é", media)
print("o maior salario ", maior)
print("A quantidade de salarios acima de 3000 é:",acima_3000)
```

```python
#2 Em uma instituição de ensino, um professor deseja analisar o desempenho de uma turma sem saber previamente a quantidade de alunos. Desenvolva um programa que leia as notas dos alunos até que seja digitado o valor -1 (fim da entrada). Ao final, calcule a média da turma, informe quantos alunos foram aprovados (nota ≥ 7) e quantos foram reprovados.
soma = 0
qtd = 0
aprovados = 0
reprovados = 0

while True:
    nota = float(input("digite a Nota (-1 para sair): "))
    if nota == -1:
        break
    soma += nota
    qtd += 1

    if nota >= 7:
        aprovados += 1
    else:
        reprovados += 1
if qtd > 0:
    media = soma / qtd
else:
    media = 0

print("Média:", media)
print("Aprovados:", aprovados)
print("Reprovados:", reprovados)
```

```python
#3 Um sistema bancário simples precisa simular operações de saque em uma conta corrente. O programa deve iniciar com um saldo informado pelo usuário e permitir saques sucessivos at/é que seja digitado 0. A cada tentativa, verifique se há saldo suficiente, atualize o saldo quando possível e informe ao usuário quando o valor exceder o saldo disponível.
saldo = float(input("Saldo inicial: "))
while True:
    saque = float(input("Valor do saque (0 para sair): "))
    if saque == 0:
        break
    if saque <= saldo:
        saldo -= saque
        print("Saque realizado. Saldo:", saldo)
    else:
        print("Saldo insuficiente!")
```

```python
#4 Uma pesquisa demográfica está sendo realizada para identificar o perfil etário de um grupo de pessoas. O programa deve ler as idades informadas até que o usuário digite 0, encerrando a coleta. Ao final, apresente a média das idades, a maior idade registrada e quantas pessoas possuem mais de 18 anos.
soma = 0
qtd = 0
maior = 0
maiores_18 = 0
while True:
    idade = int(input("Idade (0 para sair): "))
    if idade == 0:
        break
    soma += idade
    qtd += 1

    if idade > maior:
        maior = idade
    if idade > 18:
        maiores_18 += 1
media = soma / qtd
print("Média:", media)
print("Maior idade:", maior)
print("Maiores de 18:", maiores_18)
```

```python
#6 Uma loja deseja automatizar o cálculo de suas vendas diárias. Desenvolva um programa que leia o valor de 8 vendas realizadas. Sempre que o valor de uma venda for superior a R$ 100, aplique automaticamente um desconto de 10% sobre esse valor. Ao final, informe o total arrecadado após a aplicação dos descontos.
total = 0
for i in range(8):
    valor = float(input("Valor da venda: "))
    if valor > 100:
        valor *= 0.9  # 10% desconto
    total += valor
print("Total arrecadado:", total)
```

```python
# 7 Um sistema de análise numérica deve processar uma sequência de valores informados pelo usuário. O programa deve ler números inteiros até que seja digitado 0. Durante a execução, conte quantos números são pares, quantos são ímpares e calcule a soma de todos os números positivos informados.]
pares = 0
impares = 0
soma_positivos = 0
while True:
    num = int(input("Digite o número (0 para sair): "))
    if num == 0:
        break
    if num % 2 == 0:
        pares += 1
    else:
        impares += 1
    if num > 0:
        soma_positivos += num
print("Pares:", pares)
print("Ímpares:", impares)
print("Soma positivos:", soma_positivos)
```

```python
#8 Uma academia deseja monitorar o peso de seus alunos para fins estatísticos. O sistema deve ler o peso de 6 pessoas e, ao final, apresentar a média dos pesos, informar quantas pessoas possuem peso superior a 80 kg e indicar qual foi o menor peso registrado
soma = 0
acima80 = 0
for i in range(1, 7):
    peso = float(input(f"Digite o peso da {i}ª pessoa: "))
    soma += peso
    if peso > 80:
        acima80 += 1
    if i == 1:
        menor = peso
    if peso < menor:
        menor = peso
media = soma / 6
print(f"Média dos pesos: {media:.2f} kg")
print(f"Pessoas acima de 80kg: {acima80}")
print(f"Menor peso registrado: {menor:.2f} kg")
```

```python
#9 Um sistema de votação eletrônica simplificado será utilizado em uma eleição interna. Os votos devem ser informados por meio de códigos numéricos (1, 2 ou 3), sendo que o valor 0 indica o encerramento da votação. O programa deve contabilizar os votos de cada candidato e, ao final, informar qual candidato obteve a maior quantidade de votos.
voto1 = 0
voto2 = 0
voto3 = 0
voto = int(input("Digite o voto (1, 2, 3 ou 0 para encerrar): "))
while voto != 0:
    if voto == 1:
        voto1 += 1
    elif voto == 2:
        voto2 += 1
    elif voto == 3:
        voto3 += 1
    else:
        print("Voto inválido")
    voto = int(input("Digite o voto (1, 2, 3 ou 0 para encerrar): "))
print(f"Votos candidato 1: {voto1}")
print(f"Votos candidato 2: {voto2}")
print(f"Votos candidato 3: {voto3}")
if voto1 > voto2 and voto1 > voto3:
    print("Candidato 1 venceu")
elif voto2 > voto1 and voto2 > voto3:
    print("Candidato 2 venceu")
elif voto3 > voto1 and voto3 > voto2:
    print("Candidato 3 venceu")
else:
    print("Empate")
```

```python
#10 Uma indústria deseja acompanhar sua produção semanal para avaliar o desempenho operacional. O programa deve ler a quantidade produzida em cada um dos 7 dias da semana e, ao final, calcular a média de produção, informar o maior valor registrado e indicar quantos dias tiveram produção inferior a 100 unidades.
soma = 0
abaixo100 = 0
for i in range(1, 8):
    producao = int(input(f"Digite a produção do {i}° dia: "))
    soma += producao
    if producao < 100:
        abaixo100 += 1
    if i == 1:
        maior = producao
    if producao > maior:
        maior = producao
media = soma / 7
print(f"Média de produção: {media:.2f}")
print(f"Maior produção registrada: {maior}")
print(f"Dias com produção inferior a 100: {abaixo100}")
```

```python
#11 Um sistema matemático deve analisar propriedades de números inteiros informados pelo usuário. O programa deve ler valores até que seja digitado -1 e, ao final, informar quantos números são múltiplos de 3 e calcular a média apenas desses números.
soma = 0
contador = 0

while True:

    numero = int(input("Digite um número (-1 para sair): "))

    if numero == -1:
        break

    if numero % 3 == 0:
        soma += numero
        contador += 1

if contador > 0:

    media = soma / contador

    print(f"Quantidade de múltiplos de 3: {contador}")
    print(f"Média dos múltiplos de 3: {media:.2f}")

else:
    print("Nenhum múltiplo de 3 foi digitado")
```

```python
#12 Em uma escola, deseja-se analisar a altura dos alunos de uma turma. O programa deve ler a altura de 5 alunos e calcular a média geral. Em seguida, deve informar quantos alunos possuem altura acima da média calculada.
soma = 0
contador = 0
alturas = []

for i in range(1, 6):

    altura = float(input(f"Digite a altura do {i}° aluno: "))

    alturas.append(altura)

    soma += altura

media = soma / 5

for altura in alturas:

    if altura > media:
        contador += 1

print(f"Média das alturas: {media:.2f}")
print(f"Alunos acima da média: {contador}")
```

```python
#13 Um sistema financeiro deve registrar depósitos realizados em uma conta. O programa deve ler valores de depósitos até que seja informado o valor 0. Ao final, deve apresentar o total depositado, o maior valor de depósito realizado e a quantidade total de depósitos efetuados.
soma = 0
contador = 0

while True:

    deposito = float(input("Digite o valor do depósito (0 para sair): "))

    if deposito == 0:
        break

    soma += deposito
    contador += 1

    if contador == 1:
        maior = deposito

    if deposito > maior:
        maior = deposito

print(f"Total depositado: R$ {soma:.2f}")
print(f"Maior depósito: R$ {maior:.2f}")
print(f"Quantidade de depósitos: {contador}")
```
