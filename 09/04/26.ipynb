```python
#1 Usando while, exiba os números de 1 até 10, um por linha. Use uma variável de controle iniciada em 1 e incremente a cada iteração.
contador = 1
while contador<=10:
    print(contador,end=' ')
    contador+=1
```

```python
#2 Defina uma senha correta no código. Usando while, peça a senha ao usuário repetidamente até que ele acerte. A cada tentativa errada, exiba 'Senha incorreta, tente novamente.'. Ao acertar, exiba 'Acesso liberado!' e quantas tentativas foram necessárias.
contador=0
senha_correta="123"
while True:
    senha = input("digite sua senha:")
    if senha == senha_correta:
        contador += 1
        print("acesso liberado!")
        print(f"usou {contador} tentativa ")
        break
    else:
        print(f"acesso negado,tente novamente")
    contador += 1
```

```python
#3 Defina um número secreto entre 1 e 100 no código. Usando while, leia tentativas do usuário até acertar. A cada erro, diga se o número secreto é maior ou menor. Ao final, exiba quantas tentativas foram usadas.
contador = 0
numero = 52 # Corrigido: era uma anotação de tipo, agora é uma atribuição de inteiro
while True:
    tentativa = int(input("Digite sua tentativa (1 a 100): "))
    contador += 1 # Corrigido: usando 'contador' consistentemente
    if 1 <= tentativa <= 100:
        if tentativa == numero:
            print(f"Acertou! O número secreto era {numero}. Você usou {contador} tentativas.")
            break
        elif tentativa > numero:
            print("O número secreto é menor.")
        else: # tentativa < numero
            print("O número secreto é maior.")
    else:
        print("Por favor, digite um número entre 1 e 100.")
```
