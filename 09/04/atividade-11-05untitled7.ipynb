```python
#1 Crie uma tupla contendo três coordenadas no formato (x, y). Exiba cada coordenada separadamente, calcule a soma de todos os valores armazenados e informe qual é o maior número presente na tupla.
coordenadas = [(1,2),(3,4),(5,6)]
print("Coordenadas individuais:")
for i, coordenada_item in enumerate(coordenadas):
    print(f"Coordenada {i+1}: {coordenada_item}")

soma = sum(x+y for x,y in coordenadas)
print(f"Soma de todos os valores: {soma}")

# Para encontrar o maior número presente na tupla, precisamos considerar todos os valores individualmente
# Flatten a lista de tuplas em uma única lista de números
todos_valores = [valor for par_coord in coordenadas for valor in par_coord]
maior = max(todos_valores) # Encontrando o valor máximo da lista "achatada"
print(f"O maior número presente nas coordenadas é: {maior}")
```

```python
#3 Crie uma tupla com 6 números inteiros. Exiba a quantidade de números pares, a soma dos números ímpares e o menor valor armazenado.
numeros = (1,2,3,4,5,6)
soma_impares = 0
pares_count = 0
menor_valor = numeros[0] # Inicializa com o primeiro elemento para encontrar o menor

for num in numeros:
    if num % 2 == 0:
        pares_count += 1
    else:
        soma_impares += num # Calcula a soma dos números ímpares

    if num < menor_valor:
        menor_valor = num

print(f"Menor valor é: {menor_valor}")
print(f"Quantidade de pares: {pares_count}")
print(f"Soma dos valores ímpares: {soma_impares}")
```

```python
#4 Crie dois conjuntos contendo nomes de alunos de duas turmas diferentes. Exiba um novo conjunto contendo todos os alunos sem repetição usando a operação de união.
turmaA_lista = ['ze chico','joao enertos','arthur','joaquim']
turmaB_lista = ['maria',' victoria','joao', 'kaique']

# Converter as listas em conjuntos para remover duplicatas e permitir a operação de união
turmaA_set = set(turmaA_lista)
turmaB_set = set(turmaB_lista)

# Realizar a operação de união entre os dois conjuntos
turma_unida = turmaA_set.union(turmaB_set)

print("Alunos de todas as turmas (sem repetição):")
print(turma_unida)
print(f"Quantidade total de alunos únicos: {len(turma_unida)}")
```
