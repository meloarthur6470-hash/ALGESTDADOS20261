```python
#12 Crie uma lista de 5 itens. Use enumerate() para exibir: "Item X na posição Y", onde X é o valor e Y é o índice.
lista = ['alternador','sensor','correia','escapamento','borracha']
for indice, valor in enumerate(lista):
    print(f"lista: {valor} na posição: {indice}")
```
