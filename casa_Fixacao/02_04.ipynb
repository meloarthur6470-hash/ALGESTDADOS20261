```python
#1 Usando for e range(), exiba os números de 1 até 10, um por linha.
for i in range(1, 11):
    print(i)
```

```python
#2 Usando for e range(), exiba os números de 10 até 1 em ordem decrescente, um por linha.
for i in range(10, 0, -1):
    print(i)
```

```python
#3Leia um número inteiro. Usando for e range(), exiba a tabuada completa desse número, do 1 ao 10, no formato: 'N x I = R'.
n = int(input("digite um numero inteiro "))
for i in range(1,11):
    print(f"{n}*{i}={n*i}")
```

```python
#4 Leia um número inteiro N positivo. Usando for, calcule e exiba a soma de todos os números de 1 até N. Valide que N é positivo.
n= int(input("digite um numero inteiro "))
if n > 0:
    soma= 0
    for i in range(1,n+1):
        soma+= i
    print(soma)
else:
    print("numero negativo")
```

```python
#5 Leia um número inteiro N positivo. Usando for e range(), exiba todos os números pares de 0 até N. Ao final, exiba a quantidade total de pares encontrados.
n = int(input("digite um numero inteiro:"))
contador = 0
if n > 0:
    for i in range(0,n+1,2):
        print(i)
        contado+=1
    print(f" tem {contador} pares")
else:
    print("numero negativo")
```
