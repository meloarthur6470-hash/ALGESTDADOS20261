```python
#3Leia um número inteiro. Usando for e range(), exiba a tabuada completa desse número, do 1 ao 10, no formato: 'N x I = R'.
n = int(input("digite um numero inteiro "))
for i in range(1,11):
    print(f"{n}*{i}={n*i}")
```

```python
#4 Leia um número inteiro N positivo. Usando for, calcule e exiba a soma de todos os números de 1 até N. Valide que N é positivo.
n= int(input("digite um numero inteiro "))
if n > 0:
    soma= 0
    for i in range(1,n+1):
        soma+= i
    print(soma)
else:
    print("numero negativo")
```

```python
#5 Leia um número inteiro N positivo. Usando for e range(), exiba todos os números pares de 0 até N. Ao final, exiba a quantidade total de pares encontrados.
n = int(input("digite um numero inteiro:"))
contador = 0
if n > 0:
    for i in range(0,n+1,2):
        print(i)
        contado+=1
    print(f" tem {contador} pares")
else:
    print("numero negativo")
```

```python
#6 Leia um número N de alunos (positivo). Usando for, leia a nota de cada aluno, valide que está entre 0 e 10, acumule a soma e calcule a média ao final. Exiba a média com 2 casas decimais.
n = int(input("digite o número de alunos: "))
soma = 0

if n > 0:
    for i in range(n):
        nota = float(input(f"digite a nota do aluno {i+1}: "))
        while nota < 0 or nota > 10:
            print("nota invalida! digite um valor entre 0 e 10.")
            nota = float(input(f"digite a nota do aluno {i+1}: "))
        soma += nota

    media = soma / n
    print(f"media final: {media:.2f}")
else:
    print("O número de alunos deve ser positivo.")
```
