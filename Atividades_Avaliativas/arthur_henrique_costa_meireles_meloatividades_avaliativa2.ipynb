```python
#01 - Levantamento Salarial de Uma empresa de médio porte deseja realizar um levantamento sobre a remuneração de seus funcionários.Desenvolva um programa que leia os salários de 10 funcionários, um por vez, e, ao final, informe a média salarial da empresa, o maior salário registrado e quantos funcionários recebem acima de R$ 3.000.Os valores informados são válidos e positivos.

soma_salarios = 0
maior_salario = 0
funcionarios_acima_3000 = 0
total_funcionarios = 10

for i in range(1, total_funcionarios + 1):
    salario = float(input(f"Digite o salario do {i}° funcionário: R$"))

    soma_salarios += salario

if salario >= 0:
    if salario > maior_salario:
        maior_salario = salario

    if salario > 3000:
        funcionarios_acima_3000 += 1

else:
    print('valor do salario inválido!')

media_salarial = soma_salarios / total_funcionarios

print('')
print(f'A média salarial da empresa é de: R$ {media_salarial}')
print(f'O maior salário registrado foi: R${maior_salario:.2f}')
print(f'Os funcionarios que recebem acima de 3.000: {funcionarios_acima_3000} funcionarios')
```

```python
#2 Em uma instituição de ensino, um professor deseja analisar o desempenho de uma turma sem saber previamente a quantidade de alunos. Desenvolva um programa que leia as notas dos alunos até que seja digitado o valor -1 (fim da entrada). Ao final, calcule a média da turma, informe quantos alunos foram aprovados (nota ≥ 7) e quantos foram reprovados.
soma = 0
qtd = 0
aprovados = 0
reprovados = 0

while True:
    nota = float(input("digite a Nota ou digite -1 para sair: "))
    if nota == -1:
        break
    soma += nota
    qtd += 1

    if nota >= 7:
        aprovados += 1
    else:
        reprovados += 1
if qtd > 0:
    media = soma / qtd
else:
    media = 0

print("Média:", media)
print("Aprovados:", aprovados)
print("Reprovados:", reprovados)
```

```python
#3 Um sistema bancário simples precisa simular operações de saque em uma conta corrente. O programa deve iniciar com um saldo informado pelo usuário e permitir saques sucessivos at/é que seja digitado 0. A cada tentativa, verifique se há saldo suficiente, atualize o saldo quando possível e informe ao usuário quando o valor exceder o saldo disponível.
saldo = float(input("Saldo inicial: "))
while True:
    saque = float(input("Valor do saque (0 para sair): "))
    if saque == 0:
        break
    if saque <= saldo:
        saldo -= saque
        print("Saque realizado. Saldo:", saldo)
    else:
        print("Saldo insuficiente!")
```

```python
#15 Uma loja de eletrônicos deseja calcular o valor total das vendas realizadas em um período. O programa deve ler os preços dos produtos vendidos até que seja digitado 0. Para valores superiores a R$ 200, deve ser aplicado um desconto de 5%. Ao final, apresente o valor total a ser pago após os descontos.
total = 0

preco = float(input("Digite o preço do produto: "))

while preco != 0:

    if preco > 200:
        desconto = preco * 0.05
        preco = preco - desconto

    total += preco

    preco = float(input("Digite o preço do produto: "))

print(f"Valor total das vendas: R$ {total:.2f}")
```

```python
#28 Um sistema deve analisar números inteiros informados pelo usuário até que seja digitado -1. Ao final, informe quantos números são primos e calcule a soma dos números que não são primos.
quantidade_primos = 0
soma_nao_primos = 0

numero = int(input("Digite um número (-1 encerra): "))

while numero != -1:

    divisores = 0

    for i in range(1, numero + 1):

        if numero % i == 0:
            divisores += 1

    if divisores == 2:
        quantidade_primos += 1

    else:
        soma_nao_primos += numero

    numero = int(input("Digite um número (-1 encerrar): "))

print(f"Quantidade de números primos: {quantidade_primos}")
print(f"Soma dos números não primos: {soma_nao_primos}")
```

```python
# 29. Avaliação de Notas um sistema de avaliação deve ler notas até que seja digitado 0, calcular a média geral e informar quantas notas estão acima da média.

quantidade_notas = 0
notas_acima_da_media = 0
soma = 0

while True:
    notas = float(input("Digite suas notas: "))

    if notas == 0:
        break
    else:
        quantidade_notas += 1
        soma += notas
        media = soma / quantidade_notas

        if notas > media:
            notas_acima_da_media += 1
print("-"*30)
print(f"A média geral das notas acima é de {media:.2f}.")
print(f"Dentre as notas acima, a quantidade de notas acima da média é de {notas_acima_da_media} notas.")
```

```python
# 30. Cálculo de Fatorial um programa deve ler um número inteiro positivo e calcular seu fatorial utilizando repetição após isso, deve verificar se o resultado do fatorial é par ou ímpar e apresentar essa informação ao usuário.

numero = int(input("Digite un número que seja inteiro e positivo: "))

fatorial = 1

for i in range(1 , numero + 1):
    fatorial *= i
print(f"O valor {numero}! é {fatorial}")

if fatorial % 2 == 0:
    print(f"O valor {fatorial} do número {numero}! é PAR.")
else:
    print(f"O valor {fatorial} do número {numero}! é ÍMPAR.")
```
