```python

# EXERCÍCIO 01 - Classificador de Três Números
print("\nEXERCÍCIO 01 - Classificador de Três Números")
# Input validation for n1
while True:
    try:
        n1 = int(input("Digite o primeiro número: "))
        break
    except ValueError:
        print("Entrada inválida. Por favor, digite um número inteiro válido.")

# Input validation for n2
while True:
    try:
        n2 = int(input("Digite o segundo número: "))
        break
    except ValueError:
        print("Entrada inválida. Por favor, digite um número inteiro válido.")

# Input validation for n3
while True:
    try:
        n3 = int(input("Digite o terceiro número: "))
        break
    except ValueError:
        print("Entrada inválida. Por favor, digite um número inteiro válido.")

if n1 >= n2 and n1 >= n3:
    maior = n1
    if n2 >= n3:
        meio = n2
        menor = n3
    else:
        meio = n3
        menor = n2
elif n2 >= n1 and n2 >= n3:
    maior = n2
    if n1 >= n3:
        meio = n1
        menor = n3
    else:
        meio = n3
        menor = n1
else:
    maior = n3
    if n1 >= n2:
        meio = n1
        menor = n2
    else:
        meio = n2
        menor = n1

print(f"Maior: {maior}")
print(f"Meio: {meio}")
print(f"Menor: {menor}")
```

```python
# EXERCÍCIO 02 - Calculadora de Gorjeta
print("\nEXERCÍCIO 02 - Calculadora de Gorjeta")
while True:
    try:
        valor_conta = float(input("Digite o valor da conta: R$ "))
        break
    except ValueError:
        print("Entrada inválida. Por favor, digite um valor numérico válido.")

while True:
    try:
        nota = int(input("Digite a nota do serviço (1 a 5): "))
        break
    except ValueError:
        print("Entrada inválida. Por favor, digite um número inteiro válido.")

if valor_conta < 0 or nota < 1 or nota > 5:
    print("Entrada inválida.")
else:
    if nota == 5:
        percentual = 0.20
    elif nota == 4:
        percentual = 0.15
    elif nota == 3:
        percentual = 0.10
    else:
        percentual = 0.00

    gorjeta = valor_conta * percentual
    total = valor_conta + gorjeta

    print(f"Gorjeta: R$ {gorjeta:.2f}")
    print(f"Total a pagar: R$ {total:.2f}")
```

```python
#EXERCÍCIO 03 - Verificador de Ano Bissexto
print("\nEXERCÍCIO 03 - Verificador de Ano Bissexto")
# Input validation for ano
while True:
    try:
        ano = int(input("Digite um ano: "))
        break
    except ValueError:
        print("Entrada inválida. Por favor, digite um número inteiro válido.")

if ano <= 0:
    print("Ano inválido.")
else:
    if (ano % 400 == 0):
        print("Ano bissexto.")
    elif (ano % 100 == 0):
        print("Ano não bissexto.")
    elif (ano % 4 == 0):
        print("Ano bissexto.")
    else:
        print("Ano não bissexto.")
```

```python
#EXERCÍCIO 04 - Cálculo de Imposto de Renda Simplificado
print("\nEXERCÍCIO 04 - Cálculo de Imposto de Renda Simplificado")
# Input validation for salario_bruto
while True:
    try:
        salario_bruto = float(input("Digite o salário bruto mensal: R$ "))
        break
    except ValueError:
        print("Entrada inválida. Por favor, digite um valor numérico válido.")

if salario_bruto <= 0:
    print("Salário inválido.")
else:
    if salario_bruto <= 2259:
        aliquota = 0.00
    elif salario_bruto <= 2826:
        aliquota = 0.075
    elif salario_bruto <= 3751:
        aliquota = 0.15
    elif salario_bruto <= 4664:
        aliquota = 0.225
    else:
        aliquota = 0.275

    imposto = salario_bruto * aliquota
    salario_liquido = salario_bruto - imposto

    print(f"Alíquota aplicada: {aliquota * 100:.1f}%")
    print(f"Imposto: R$ {imposto:.2f}")
    print(f"Salário líquido: R$ {salario_liquido:.2f}")
```

```python

# EXERCÍCIO 05 - Placar de Futebol
print("\nEXERCÍCIO 05 - Placar de Futebol")
# Input validation for gols_a
while True:
    try:
        gols_a = int(input("Gols do time A: "))
        break
    except ValueError:
        print("Entrada inválida. Por favor, digite um número inteiro válido.")

# Input validation for gols_b
while True:
    try:
        gols_b = int(input("Gols do time B: "))
        break
    except ValueError:
        print("Entrada inválida. Por favor, digite um número inteiro válido.")

if gols_a < 0 or gols_b < 0:
    print("Placar inválido.")
else:
    if gols_a > gols_b:
        diferenca = gols_a - gols_b
        print("Time A venceu.")
        print(f"Diferença de gols: {diferenca}")
        if diferenca > 3:
            print("Foi goleada.")
    elif gols_b > gols_a:
        diferenca = gols_b - gols_a
        print("Time B venceu.")
        print(f"Diferença de gols: {diferenca}")
        if diferenca > 3:
            print("Foi goleada.")
    else:
        print("Empate.")
```

```python
# EXERCÍCIO 06 - Validador de Data
print("\nEXERCÍCIO 06 - Validador de Data")
# Input validation for dia
while True:
    try:
        dia = int(input("Digite o dia: "))
        break
    except ValueError:
        print("Entrada inválida. Por favor, digite um número inteiro válido.")

# Input validation for mes
while True:
    try:
        mes = int(input("Digite o mês: "))
        break
    except ValueError:
        print("Entrada inválida. Por favor, digite um número inteiro válido.")

# Input validation for ano
while True:
    try:
        ano = int(input("Digite o ano: "))
        break
    except ValueError:
        print("Entrada inválida. Por favor, digite um número inteiro válido.")

if ano <= 0:
    print("Ano inválido.")
elif mes < 1 or mes > 12:
    print("Mês inválido.")
else:
    if (ano % 400 == 0) or ((ano % 4 == 0) and (ano % 100 != 0)):
        bissexto = True
    else:
        bissexto = False

    if mes in [4, 6, 9, 11]:
        max_dia = 30
    elif mes in [1, 3, 5, 7, 8, 10, 12]:
        max_dia = 31
    else:
        if bissexto:
            max_dia = 29
        else:
            max_dia = 28

    if dia < 1 or dia > max_dia:
        print("Dia inválido para o mês informado.")
    else:
        print("Data válida.")
```

```python
# EXERCÍCIO 07 - Calculadora de Frete
print("\nEXERCÍCIO 07 - Calculadora de Frete")
# Input validation for peso
while True:
    try:
        peso = float(input("Digite o peso do pacote (kg): "))
        break
    except ValueError:
        print("Entrada inválida. Por favor, digite um valor numérico válido.")

# Input validation for regiao
while True:
    try:
        regiao = int(input("Digite a região (1=Sul/Sudeste, 2=Centro-Oeste/Nordeste, 3=Norte): "))
        break
    except ValueError:
        print("Entrada inválida. Por favor, digite um número inteiro válido.")

if peso <= 0 or regiao < 1 or regiao > 3:
    print("Entrada inválida.")
else:
    if peso <= 1:
        frete_base = 15
    elif peso <= 5:
        frete_base = 25
    else:
        frete_base = 40

    if regiao == 2 or regiao == 3:
        frete_final = frete_base * 1.20
    else:
        frete_final = frete_base

    print(f"Frete final: R$ {frete_final:.2f}")
```

```python
 #EXERCÍCIO 08 - Simulador de Caixa Eletrônico
print("\nEXERCÍCIO 08 - Simulador de Caixa Eletrônico")
saldo_correto = 2000.00
senha_correta = "1234"

# Input validation for valor_saque
while True:
    try:
        valor_saque = float(input("Digite o valor do saque: R$ "))
        break
    except ValueError:
        print("Entrada inválida. Por favor, digite um valor numérico válido.")
senha_digitada = input("Digite a senha: ")

if valor_saque <= 0:
    print("Valor de saque inválido.")
else:
    if senha_digitada != senha_correta:
        print("Senha incorreta.")
    else:
        if valor_saque > saldo_correto:
            print("Saldo insuficiente.")
        else:
            saldo_correto -= valor_saque
            print(f"Saque realizado com sucesso. Saldo restante: R$ {saldo_correto:.2f}")
```

```python
# EXERCÍCIO 09 - Conversor de Temperatura Completo
print("\nEXERCÍCIO 09 - Conversor de Temperatura Completo")
# Input validation for temperatura
while True:
    try:
        temperatura = float(input("Digite a temperatura: "))
        break
    except ValueError:
        print("Entrada inválida. Por favor, digite um valor numérico válido.")
unidade = input('Digite a unidade de origem ("C", "F" ou "K"): ').upper()

match unidade:
    case "C":
        temp_celsius = temperatura
        temp_fahrenheit = (temp_celsius * 9/5) + 32
        temp_kelvin = temp_celsius + 273.15
        print(f"Fahrenheit: {temp_fahrenheit:.2f}")
        print(f"Kelvin: {temp_kelvin:.2f}")
    case "F":
        temp_fahrenheit = temperatura
        temp_celsius = (temp_fahrenheit - 32) * 5/9
        temp_kelvin = temp_celsius + 273.15
        print(f"Celsius: {temp_celsius:.2f}")
        print(f"Kelvin: {temp_kelvin:.2f}")
    case "K":
        if temperatura < 0:
            print("Temperatura em Kelvin não pode ser negativa.")
        else:
            temp_kelvin = temperatura
            temp_celsius = temp_kelvin - 273.15
            temp_fahrenheit = (temp_celsius * 9/5) + 32
            print(f"Celsius: {temp_celsius:.2f}")
            print(f"Fahrenheit: {temp_fahrenheit:.2f}")
    case _:
        print("Unidade inválida.")
```

```python
# EXERCÍCIO 10 - Classificador de IMC com Risco
print("\nEXERCÍCIO 10 - Classificador de IMC com Risco")
# Input validation for peso
while True:
    try:
        peso = float(input("Digite o peso (kg): "))
        break
    except ValueError:
        print("Entrada inválida. Por favor, digite um valor numérico válido.")

# Input validation for altura
while True:
    try:
        altura = float(input("Digite a altura (m): "))
        break
    except ValueError:
        print("Entrada inválida. Por favor, digite um valor numérico válido.")

if peso <= 0 or altura < 0.5 or altura > 2.5:
    print("Valores inválidos.")
else:
    imc = peso / (altura ** 2)

    if imc < 18.5:
        classificacao = "Abaixo do peso"
        risco = "baixo"
    elif imc < 25:
        classificacao = "Normal"
        risco = "baixo"
    elif imc < 30:
        classificacao = "Sobrepeso"
        risco = "moderado"
    elif imc < 35:
        classificacao = "Obesidade I"
        risco = "moderado"
    elif imc < 40:
        classificacao = "Obesidade II"
        risco = "alto"
    else:
        classificacao = "Obesidade III"
        risco = "alto"

    print(f"IMC: {imc:.2f}")
    print(f"Classificação: {classificacao}")
    print(f"Risco cardiovascular: {risco}")
```

```python
# EXERCÍCIO 11 - Detector de Múltiplos
print("\nEXERCÍCIO 11 - Detector de Múltiplos")
# Input validation for numero
while True:
    try:
        numero = int(input("Digite um número inteiro positivo: "))
        break
    except ValueError:
        print("Entrada inválida. Por favor, digite um número inteiro válido.")

if numero <= 0:
    print("Número inválido.")
else:
    eh_multiplo_de_2 = numero % 2 == 0
    eh_multiplo_de_3 = numero % 3 == 0
    eh_multiplo_de_5 = numero % 5 == 0

    if eh_multiplo_de_2:
        print("É múltiplo de 2.")
    if eh_multiplo_de_3:
        print("É múltiplo de 3.")
    if eh_multiplo_de_5:
        print("É múltiplo de 5.")
    if not eh_multiplo_de_2 and not eh_multiplo_de_3 and not eh_multiplo_de_5:
        print("Não é múltiplo de 2, 3 ou 5.")

    if eh_multiplo_de_2 and eh_multiplo_de_3 and eh_multiplo_de_5:
        print("É múltiplo de 2, 3 e 5 ao mesmo tempo.")
```

```python
# EXERCÍCIO 12 - Calculadora de Desconto Progressivo
print("\nEXERCÍCIO 12 - Calculadora de Desconto Progressivo")
# Input validation for preco_original
while True:
    try:
        preco_original = float(input("Digite o preço original do produto: R$ "))
        break
    except ValueError:
        print("Entrada inválida. Por favor, digite um valor numérico válido.")
cupom = input('Digite o cupom ("PROMO10", "PROMO20", "PROMO30" ou "SEMCUPOM"): ').upper()

if preco_original <= 0:
    print("Preço inválido.")
else:
    match cupom:
        case "PROMO10":
            desconto_percentual = 0.10
        case "PROMO20":
            desconto_percentual = 0.20
        case "PROMO30":
            desconto_percentual = 0.30
        case "SEMCUPOM":
            desconto_percentual = 0.00
        case _:
            desconto_percentual = 0.00
            print("Cupom inválido. Considerando sem desconto.")

    if preco_original < 50 and desconto_percentual > 0.10:
        desconto_percentual = 0.10

    desconto_reais = preco_original * desconto_percentual
    preco_final = preco_original - desconto_reais

    print(f"Cupom usado: {cupom}")
    print(f"Desconto: R$ {desconto_reais:.2f}")
    print(f"Preço final: R$ {preco_final:.2f}")
```

```python
# EXERCÍCIO 13 - Avaliador de Triângulo Completo
print("\nEXERCÍCIO 13 - Avaliador de Triângulo Completo")
# Input validation for lado1
while True:
    try:
        lado1 = float(input("Digite o primeiro lado: "))
        break
    except ValueError:
        print("Entrada inválida. Por favor, digite um valor numérico válido.")

# Input validation for lado2
while True:
    try:
        lado2 = float(input("Digite o segundo lado: "))
        break
    except ValueError:
        print("Entrada inválida. Por favor, digite um valor numérico válido.")

# Input validation for lado3
while True:
    try:
        lado3 = float(input("Digite o terceiro lado: "))
        break
    except ValueError:
        print("Entrada inválida. Por favor, digite um valor numérico válido.")

if lado1 <= 0 or lado2 <= 0 or lado3 <= 0:
    print("Lados inválidos.")
else:
    if lado1 < lado2 + lado3 and lado2 < lado1 + lado3 and lado3 < lado1 + lado2:
        if lado1 == lado2 == lado3:
            print("Triângulo equilátero.")
        elif lado1 == lado2 or lado1 == lado3 or lado2 == lado3:
            print("Triângulo isósceles.")
        else:
            print("Triângulo escaleno.")
    else:
        print("Os lados não formam um triângulo.")
```

```python
# EXERCÍCIO 14 - Calculadora de Notas com Conceito
print("\nEXERCÍCIO 14 - Calculadora de Notas com Conceito")
# Input validation for nota1
while True:
    try:
        nota1 = float(input("Digite a 1ª nota: "))
        break
    except ValueError:
        print("Entrada inválida. Por favor, digite um valor numérico válido.")

# Input validation for nota2
while True:
    try:
        nota2 = float(input("Digite a 2ª nota: "))
        break
    except ValueError:
        print("Entrada inválida. Por favor, digite um valor numérico válido.")

# Input validation for nota3
while True:
    try:
        nota3 = float(input("Digite a 3ª nota: "))
        break
    except ValueError:
        print("Entrada inválida. Por favor, digite um valor numérico válido.")

# Input validation for nota4
while True:
    try:
        nota4 = float(input("Digite a 4ª nota: "))
        break
    except ValueError:
        print("Entrada inválida. Por favor, digite um valor numérico válido.")

if (nota1 < 0 or nota1 > 10 or
    nota2 < 0 or nota2 > 10 or
    nota3 < 0 or nota3 > 10 or
    nota4 < 0 or nota4 > 10):
    print("Notas inválidas.")
else:
    media = (nota1 + nota2 + nota3 + nota4) / 4

    if media >= 9:
        conceito = "A"
        status = "Aprovado"
    elif media >= 7:
        conceito = "B"
        status = "Aprovado"
    elif media >= 5:
        conceito = "C"
        status = "Recuperação"
    elif media >= 3:
        conceito = "D"
        status = "Reprovado"
    else:
        conceito = "E"
        status = "Reprovado"

    print(f"Média: {media:.2f}")
    print(f"Conceito: {conceito}")
    print(f"Status: {status}")
```

```python
# EXERCÍCIO 15 - Verificador de Acesso por Horário
print("\nEXERCÍCIO 15 - Verificador de Acesso por Horário")
# Input validation for hora
while True:
    try:
        hora = int(input("Digite a hora atual (0 a 23): "))
        break
    except ValueError:
        print("Entrada inválida. Por favor, digite um número inteiro válido.")
tipo_usuario = input('Digite o tipo de usuário ("admin", "operador", "visitante"): ').lower()

if hora < 0 or hora > 23:
    print("Hora inválida.")
else:
    if 6 <= hora <= 11:
        turno = "manhã"
    elif 12 <= hora <= 17:
        turno = "tarde"
    elif 18 <= hora <= 23:
        turno = "noite"
    else:
        turno = "madrugada"

    match tipo_usuario:
        case "admin":
            acesso = True
        case "operador":
            acesso = turno == "manhã" or turno == "tarde"
        case "visitante":
            acesso = turno == "tarde"
        case _:
            acesso = False
            print("Tipo de usuário inválido.")

    if tipo_usuario in ["admin", "operador", "visitante"]:
        print(f"Turno atual: {turno}")
        if acesso:
            print("Acesso permitido.")
        else:
            print("Acesso negado.")
```

```python
# EXERCÍCIO 16 - Simulador de Bomba de Combustível
print("\nEXERCÍCIO 16 - Simulador de Bomba de Combustível")
tipo_combustivel = input('Digite o combustível ("gasolina", "etanol", "diesel"): ').lower()
# Input validation for litros
while True:
    try:
        litros = float(input("Digite a quantidade de litros: "))
        break
    except ValueError:
        print("Entrada inválida. Por favor, digite um valor numérico válido.")

preco_gasolina = 5.79
preco_etanol = 3.89
preco_diesel = 6.19

if litros <= 0:
    print("Quantidade inválida.")
else:
    match tipo_combustivel:
        case "gasolina":
            preco_litro = preco_gasolina
        case "etanol":
            preco_litro = preco_etanol
        case "diesel":
            preco_litro = preco_diesel
        case _:
            preco_litro = None

    if preco_litro is None:
        print("Tipo de combustível inválido.")
    else:
        subtotal = litros * preco_litro

        if litros > 40:
            desconto = subtotal * 0.03
        else:
            desconto = 0

        total = subtotal - desconto

        print(f"Subtotal: R$ {subtotal:.2f}")
        print(f"Desconto: R$ {desconto:.2f}")
        print(f"Total a pagar: R$ {total:.2f}")
```

```python
# EXERCÍCIO 17 - Classificador de Velocidade
print("\nEXERCÍCIO 17 - Classificador de Velocidade")
# Input validation for velocidade
while True:
    try:
        velocidade = float(input("Digite a velocidade em km/h: "))
        break
    except ValueError:
        print("Entrada inválida. Por favor, digite um valor numérico válido.")
tipo_via = input('Digite o tipo da via ("urbana", "rodovia", "autopista"): ').lower()

if velocidade < 0:
    print("Velocidade inválida.")
else:
    match tipo_via:
        case "urbana":
            limite = 60
        case "rodovia":
            limite = 100
        case "autopista":
            limite = 120
        case _:
            limite = None

    if limite is None:
        print("Tipo de via inválido.")
    else:
        excesso = velocidade - limite

        if velocidade <= limite:
            situacao = "Dentro do limite"
        elif velocidade <= limite * 1.20:
            situacao = "Excesso leve"
        elif velocidade <= limite * 1.50:
            situacao = "Excesso moderado"
        else:
            situacao = "Excesso grave"

        print(f"Limite da via: {limite} km/h")
        print(f"Velocidade informada: {velocidade:.2f} km/h")
        print(f"Situação: {situacao}")
```

```python
# EXERCÍCIO 18 - Calculadora de Parcelas com Juros
print("\nEXERCÍCIO 18 - Calculadora de Parcelas com Juros")
# Input validation for valor_compra
while True:
    try:
        valor_compra = float(input("Digite o valor da compra: R$ "))
        break
    except ValueError:
        print("Entrada inválida. Por favor, digite um valor numérico válido.")

# Input validation for parcelas
while True:
    try:
        parcelas = int(input("Digite o número de parcelas (1 a 12): "))
        break
    except ValueError:
        print("Entrada inválida. Por favor, digite um número inteiro válido.")

if valor_compra <= 0 or parcelas < 1 or parcelas > 12:
    print("Entrada inválida.")
else:
    if parcelas == 1:
        juros = 0.00
    elif parcelas <= 3:
        juros = 0.05
    elif parcelas <= 6:
        juros = 0.10
    else:
        juros = 0.15

    valor_total = valor_compra * (1 + juros)
    valor_parcela = valor_total / parcelas

    print(f"Valor total com juros: R$ {valor_total:.2f}")
    print(f"Valor de cada parcela: R$ {valor_parcela:.2f}")

    if valor_parcela < 20:
        print("Aviso: parcela mínima de R$20,00 não atingida.")
        parcelas_minimas = int(valor_total // 20)
        if valor_total % 20 != 0:
            parcelas_minimas += 1
        print(f"Número mínimo de parcelas para respeitar R$20,00: {parcelas_minimas}")
```

```python
# EXERCÍCIO 19 - Detector de Intervalo com Tolerância
print("\nEXERCÍCIO 19 - Detector de Intervalo com Tolerância")
# Input validation for alvo
while True:
    try:
        alvo = float(input("Digite o valor alvo: "))
        break
    except ValueError:
        print("Entrada inválida. Por favor, digite um valor numérico válido.")

# Input validation for tolerancia
while True:
    try:
        tolerancia = float(input("Digite a tolerância: "))
        break
    except ValueError:
        print("Entrada inválida. Por favor, digite um valor numérico válido.")

# Input validation for valor_medido
while True:
    try:
        valor_medido = float(input("Digite o valor medido: "))
        break
    except ValueError:
        print("Entrada inválida. Por favor, digite um valor numérico válido.")

if alvo <= 0 or tolerancia < 0 or valor_medido <= 0:
    print("Entrada inválida.")
else:
    limite_inferior = alvo - tolerancia
    limite_superior = alvo + tolerancia
    desvio = valor_medido - alvo

    if limite_inferior <= valor_medido <= limite_superior:
        classificacao = "Dentro da faixa"
    elif valor_medido < limite_inferior:
        classificacao = "Abaixo do limite"
    else:
        classificacao = "Acima do limite"

    print(f"Limite inferior: {limite_inferior:.2f}")
    print(f"Limite superior: {limite_superior:.2f}")
    print(f"Desvio: {desvio:.2f}")
    print(f"Classificação: {classificacao}")
```

```python
# EXERCÍCIO 22 - Verificador de Intervalo Múltiplo
# ============================================================
print("\nEXERCÍCIO 22 - Verificador de Intervalo Múltiplo")
# Input validation for A
while True:
    try:
        A = float(input("Digite o intervalo mínimo A: "))
        break
    except ValueError:
        print("Entrada inválida. Por favor, digite um valor numérico válido.")

# Input validation for B
while True:
    try:
        B = float(input("Digite o intervalo máximo B: "))
        break
    except ValueError:
        print("Entrada inválida. Por favor, digite um valor numérico válido.")

# Input validation for N
while True:
    try:
        N = int(input("Digite o número N: "))
        break
    except ValueError:
        print("Entrada inválida. Por favor, digite um número inteiro válido.")

# Input validation for D
while True:
    try:
        D = int(input("Digite o divisor D: "))
        break
    except ValueError:
        print("Entrada inválida. Por favor, digite um número inteiro válido.")

if A >= B or D == 0 or D < 0:
    print("Entrada inválida.")
else:
    dentro_intervalo = A <= N <= B
    divisivel = N % D == 0

    if dentro_intervalo and divisivel:
        print("Dentro do intervalo e divisível.")
    elif dentro_intervalo and not divisivel:
        print("Dentro do intervalo e não divisível.")
    elif not dentro_intervalo and divisivel:
        print("Fora do intervalo e divisível.")
    else:
        print("Fora do intervalo e não divisível.")
```

```python
# EXERCÍCIO 24 - Calculadora de Reajuste Salarial
print("\nEXERCÍCIO 24 - Calculadora de Reajuste Salarial")
# Input validation for salario_atual
while True:
    try:
        salario_atual = float(input("Digite o salário atual: R$ "))
        break
    except ValueError:
        print("Entrada inválida. Por favor, digite um valor numérico válido.")

# Input validation for avaliacao
while True:
    try:
        avaliacao = float(input("Digite o percentual de avaliação de desempenho (0 a 100): "))
        break
    except ValueError:
        print("Entrada inválida. Por favor, digite um valor numérico válido.")

if salario_atual <= 0 or avaliacao < 0 or avaliacao > 100:
    print("Entrada inválida.")
else:
    if avaliacao >= 90:
        percentual = 0.15
    elif avaliacao >= 75:
        percentual = 0.10
    elif avaliacao >= 60:
        percentual = 0.05
    else:
        percentual = 0.00

    reajuste = salario_atual * percentual
    novo_salario = salario_atual + reajuste

    if novo_salario > 15000:
        novo_salario = 15000
        reajuste = novo_salario - salario_atual
        print("Aviso: reajuste limitado ao teto de R$15.000,00.")

    print(f"Salário anterior: R$ {salario_atual:.2f}")
    print(f"Percentual aplicado: {percentual * 100:.1f}%")
    print(f"Reajuste em reais: R$ {reajuste:.2f}")
    print(f"Novo salário: R$ {novo_salario:.2f}")
```

```python
# EXERCÍCIO 29 - Simulador de Batalha por Tipos
print("\nEXERCÍCIO 29 - Simulador de Batalha por Tipos")
tipo_atacante = input('Digite o tipo do atacante ("fogo", "agua", "planta", "eletrico", "terra"): ').lower()
tipo_defensor = input('Digite o tipo do defensor ("fogo", "agua", "planta", "eletrico", "terra"): ').lower()
# Input validation for poder_base
while True:
    try:
        poder_base = float(input("Digite o poder base do ataque: "))
        break
    except ValueError:
        print("Entrada inválida. Por favor, digite um valor numérico válido.")

tipos_validos = ["fogo", "agua", "planta", "eletrico", "terra"]

if tipo_atacante not in tipos_validos or tipo_defensor not in tipos_validos or poder_base <= 0:
    print("Entrada inválida.")
else:
    vantagem = (
        (tipo_atacante == "fogo" and tipo_defensor == "planta") or
        (tipo_atacante == "agua" and tipo_defensor == "fogo") or
        (tipo_atacante == "planta" and tipo_defensor == "agua") or
        (tipo_atacante == "eletrico" and tipo_defensor == "agua") or
        (tipo_atacante == "terra" and tipo_defensor == "eletrico")
         )

    desvantagem = (
        (tipo_defensor == "fogo" and tipo_atacante == "planta") or
        (tipo_defensor == "agua" and tipo_atacante == "fogo") or
        (tipo_defensor == "planta" and tipo_atacante == "agua") or
        (tipo_defensor == "eletrico" and tipo_atacante == "agua") or
        (tipo_defensor == "terra" and tipo_atacante == "eletrico")
    )

    if vantagem:
        multiplicador = 2
        resultado = "vantagem"
    elif desvantagem:
        multiplicador = 0.5
        resultado = "desvantagem"
    else:
        multiplicador = 1
        resultado = "neutro"

    dano_final = poder_base * multiplicador

    print(f"Situação: {resultado}")
    print(f"Dano final: {dano_final:.2f}")
```
