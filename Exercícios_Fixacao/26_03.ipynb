```python
#1 Leia a nota de um aluno. Classifique usando elif: nota ≥ 9 → "A — Excelente", ≥ 7 → "B — Aprovado", ≥ 5 → "C — Recuperação", abaixo → "D — Reprovado".
nota = float(input( "digite sua nota:"))
if nota >= 9:
    print("A - Excelente")
elif nota >= 7:
    print("B - Aprovado")
elif nota >= 5:
    print("C - recuperação")
else:
    print("D - reprovado ")
```

```python
 #2 nome do dia correspondente (1 = Segunda, ..., 7 = Domingo). Para qualquer outro número, exiba "Número inválido".
 dia = int(input("digite o numero de 1 a 7: "))
match dia:
    case 1: print("Segunda feira")
    case 2: print("Terça feira")
    case 3: print("Quarta feira")
    case 4: print("Quinta feira")
    case 5: print("Sexta feira")
    case 6: print("Sabado")
    case 7: print("Domigo")
    case _:  print(" Dia nao existente!")
```

```python
 #3 Leia uma nota. Valide se está entre 0 e 10 usando encadeamento de comparações (0 <= nota <= 10). Se inválida, exiba "Nota inválida!". Se válida, exiba "Nota registrada: [nota]".
nota = float(input("digite sua nota:"))
if 0 <= nota <=10:
    print(f"Nota registrada:{nota}")
    print("nota valida")
else:
    print("nota invalida")


```

```python
#4 Leia a idade. Classifique: menor de 12 → "Criança", 12 a 17 → "Adolescente", 18 a 59 → "Adulto", 60 ou mais → "Idoso". Valide que a idade não é negativa.
idade = float(input(" digite sua idade:"))
if idade <0 or idade >=120:
    print("idade invalidada ")
elif idade >=1 and idade < 12:
    print("Criança")
elif idade >=12 and idade <18:
    print("Adolecente")
elif idade >=18 and idade <60:
    print("adulto")
elif idade >=60:
    print("idoso")
```
