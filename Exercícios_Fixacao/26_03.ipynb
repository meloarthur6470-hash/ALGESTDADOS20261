```python
#1 Leia a nota de um aluno. Classifique usando elif: nota ≥ 9 → "A — Excelente", ≥ 7 → "B — Aprovado", ≥ 5 → "C — Recuperação", abaixo → "D — Reprovado".
nota = float(input( "digite sua nota:"))
if nota >= 9:
    print("A - Excelente")
elif nota >= 7:
    print("B - Aprovado")
elif nota >= 5:
    print("C - recuperação")
else:
    print("D - reprovado ")
```

```python
 #2 nome do dia correspondente (1 = Segunda, ..., 7 = Domingo). Para qualquer outro número, exiba "Número inválido".
 dia = int(input("digite o numero de 1 a 7: "))
match dia:
    case 1: print("Segunda feira")
    case 2: print("Terça feira")
    case 3: print("Quarta feira")
    case 4: print("Quinta feira")
    case 5: print("Sexta feira")
    case 6: print("Sabado")
    case 7: print("Domigo")
    case _:  print(" Dia nao existente!")
```
